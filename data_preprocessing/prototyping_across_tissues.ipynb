```python
import pandas as pd
import numpy as np
from glob import glob
import pickle
```

```python
clustering_results_folder = ''

cluster_centers = []
for item in glob(f'{clustering_results_folder}/k_16/*/*centroids_image.npy'):
    cluster_centers.extend(np.load(item)[0])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=16, random_state=0)
kmeans.fit(cluster_centers)
new_centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
df_list = []
cluster_centers = []
all_centroids = []
reassigned_clusters = []
for i, item in enumerate(glob(f'{clustering_results_folder}/k_16/*/*centroids_image.npy')):
    df = pd.read_csv(item.replace('_centroids_image.npy', '_coords_with_clusters.csv'))
    reassigned_dict = {j: labels[i*16:(i+1)*16][j] for j in range(16)}
    df['image_prototype'] = df['image_prototype'].map(reassigned_dict)
    df_list.append(df)
    reassigned_clusters.extend(list(reassigned_dict.values()))
    
    all_centroids.extend(np.load(item)[0])
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
combined_df.to_csv(f'{clustering_results_folder}/_updated_coords_hest_patching_prototypes16_90_10_split_2step.csv')
```
